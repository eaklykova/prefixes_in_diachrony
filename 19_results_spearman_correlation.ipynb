```python
import pandas as pd
from scipy.stats import spearmanr
```

#### Получение списка глаголов

```python
rank_df = pd.read_csv('prefixes_ranked.tsv', sep='\t')
rank_df
```

```python
pre_to_sov_df = rank_df[list(rank_df.columns)[0:6]]
for col in list(rank_df.columns)[0:6]:
    pre_to_sov_df[col.split('_')[0]] = pre_to_sov_df[col].astype('category').cat.codes
pre_to_sov_df.drop(columns=list(rank_df.columns)[0:6], inplace=True)
```

```python
pre_to_sov_df.corr(method='spearman')
```

```python
sov_to_post_df = rank_df[list(rank_df.columns)[6:12]]
for col in list(rank_df.columns)[6:12]:
    sov_to_post_df[col.split('_')[0]] = sov_to_post_df[col].astype('category').cat.codes
sov_to_post_df.drop(columns=list(rank_df.columns)[6:12], inplace=True)
```

```python
pre_to_post_df = rank_df[list(rank_df.columns)[12:18]]
for col in list(rank_df.columns)[12:18]:
    pre_to_post_df[col.split('_')[0]] = pre_to_post_df[col].astype('category').cat.codes
pre_to_post_df.drop(columns=list(rank_df.columns)[12:18], inplace=True)
```

```python
((pre_to_sov_df.corr(method='spearman') + sov_to_post_df.corr(method='spearman') + pre_to_post_df.corr(method='spearman')) / 3).round(3)
```
